```python
import random
import time
import sys

# 🔮 MYSTICAL NUMBER SYSTEM (Each color has its own sacred numbers)
predictions = {
    "yellow": {3, 12, 21, 33, 48},   # Enlightenment & Success ☀️
    "green": {7, 16, 22, 37, 45},   # Growth & Luck 🍀
    "blue": {5, 11, 19, 28, 42},   # Wisdom & Destiny 🌊
    "red": {9, 14, 27, 35, 50}     # Power & Passion 🔥
}

# 📜 Stores all fortunes received by the player
history = []

# 🔥 7% CHANCE FOR A LEGENDARY PROPHECY
rare_fortune = [
    "🌌✨ The universe has chosen you for a legendary destiny! 🚀",
    "💎🏆 You will leave behind a legacy remembered for generations!",
    "⚡🌠 A great power within you is awakening!"
]

# 🎭 MYSTICAL LOADING ANIMATION
def countdown_timer():
    """Creates a truly mystical waiting effect with varied speeds and a cinematic feel."""
    
    messages = [
        "✨ The energies are shifting...",
        "🔮 The Oracle is focusing...",
        "🌌 The stars are aligning...",
        "🕯️ The spirits are whispering...",
        "🌠 A great vision is forming..."
    ]

    # 🌙 Display each mystical message with a smooth transition
    for msg in messages:
        print(f"\r{msg}    ", end="", flush=True)  
        time.sleep(random.uniform(2, 3))  

    print("\n🔮 Consulting the celestial energies...", end="", flush=True)

    for _ in range(random.randint(8, 12)):  
        time.sleep(random.uniform(1, 2))  
        print(".", end="", flush=True)

    time.sleep(3)  
    print("\n✨ The vision is clear! ✨\n")  

# 🔢 SECRET NUMBER REVEAL FUNCTION (WITH HIDDEN MEANINGS!)
def reveal_numbers(color):
    """Mystically reveals the available numbers for the chosen color, with deeper meanings!"""
    time.sleep(1)
    print("\n📜 The spirits are whispering sacred numbers to you...")
    time.sleep(2)

    meanings = {
        3: "🔆 Divine creativity is calling you...",
        5: "🌊 A powerful transformation is coming...",
        7: "🍀 Luck and mystery intertwine...",
        9: "🔥 Power and leadership await...",
        11: "🧠 Wisdom will guide your path...",
        12: "📜 A hidden truth will be revealed...",
        14: "⚡ Energy and passion flow through you...",
        16: "🔮 Destiny is shifting in your favor...",
        19: "🌙 The past and future are converging...",
        21: "🏛 Ancient knowledge surrounds you...",
        22: "🤝 A fateful partnership will arise...",
        27: "⏳ The universe is testing your resolve...",
        28: "🚀 A journey to the unknown begins...",
        33: "🌟 You are chosen for something greater...",
        35: "🦾 Strength and courage are your allies...",
        37: "🔗 A connection to the past is resurfacing...",
        42: "💫 The ultimate answer is within you...",
        45: "📩 A message from beyond is coming...",
        48: "👁 Your intuition is at its peak...",
        50: "🔁 A cycle is closing, and a new one begins..."
    }

    for num in sorted(predictions[color]):  
        print(f"✨ {num} ✨ - {meanings[num]}", flush=True)
        time.sleep(random.uniform(2, 3))  

    print("\n")

# 📝 FUNCTION TO GET A RANDOM FORTUNE
def get_prediction(color, number):
    """Returns a specific fortune based on the chosen color and number.
       Includes a 7% chance to receive a legendary prophecy."""
    if random.randint(1, 100) <= 7:  
        fortune = random.choice(rare_fortune)
    else:
        meaning = {
            3: "🎭 An unexpected event will change everything!",
            5: "🌀 A path you never considered will open...",
            7: "💰 An opportunity for wealth is near!",
            9: "🦁 Your strength will be tested soon...",
            11: "🔑 A long-kept secret will come to light...",
            12: "📖 A powerful lesson will shape your future...",
            14: "🔥 Passion and energy will fuel your next move!",
            16: "✨ Your destiny is shifting—pay attention!",
            19: "🌙 A powerful transformation awaits you...",
            21: "🕊 A sign from the universe will reveal itself...",
            22: "🤝 A fateful meeting will change your course!",
            27: "🔄 The past is not finished with you yet...",
            28: "🌍 Your journey is leading to new horizons...",
            33: "🌟 A rare blessing is about to enter your life!",
            35: "⚔️ A great challenge will prove your strength!",
            37: "💭 Your dreams hold a message—listen closely!",
            42: "💫 The answer you seek is already within you...",
            45: "📜 A prophecy written long ago concerns you...",
            48: "🔮 Trust your intuition—it is rarely wrong...",
            50: "🌀 The cycle is ending, prepare for renewal..."
        }
        fortune = f"🔮 The number {number} reveals: {meaning[number]}"  
    history.append(f"{fortune}")  
    return fortune

# 🛡️ FUNCTION TO HANDLE USER INPUT SAFELY
def ask_question(prompt, options):
    """Ensures that the user enters a valid choice from the given options."""
    while True:
        answer = input(prompt).strip().lower()
        if answer in options:
            return answer  
        print("❌ Invalid choice! Please select from:", options)

# 📜 FINAL SUMMARY OF THE GAME SESSION
def display_final_summary(colors_chosen):
    """Displays a beautifully formatted summary of the game session."""
    print("\n━━━━━━━━━━━━ 📜 FINAL DESTINY SUMMARY 📜 ━━━━━━━━━━━━")
    print("🔮 Your Fortunes:")
    for i, h in enumerate(history, 1):
        print(f"  {i}️⃣ {h}")
    print("✨ The stars have spoken... until next time! 🌙✨")

# 🎭 MAIN FUNCTION THAT RUNS THE FORTUNE TELLER GAME
def fortune_teller():
    """Runs the interactive fortune teller game where users select a color and number
       to receive a deeply symbolic prophecy. Includes a countdown timer and a final session summary."""
    
    print("\n━━━━━━━━━━━━ 🔮 WELCOME TO THE MYSTIC ORACLE 🔮 ━━━━━━━━━━━━")
    time.sleep(2)
    print("✨ I am the ancient seer, and I will reveal your future... ✨\n")

    colors_chosen = []  

    while True:
        # Step 1: User selects a color
        color = ask_question("🎨 Choose a color [yellow, green, blue, red]: ", predictions.keys())
        colors_chosen.append(color)  

        # Step 2: Reveal the mysterious numbers before choosing
        reveal_numbers(color)  

        # Step 3: User selects a number
        number = int(ask_question("\n🔢 Choose your number: ", [str(n) for n in predictions[color]]))

        # Step 4: Mystic Animation
        countdown_timer()  

        # Step 5: Display the fortune
        print(f"\n📜 **Your Divine Fortune:** {get_prediction(color, number)} 🔥\n")

        # Step 6: Ask if the user wants to play again
        if ask_question("🔄 Would you like another reading? (y/n): ", ["y", "n"]) == "n":
            print("\n✨ Analyzing your spiritual aura... 🧘")
            time.sleep(3)
            display_final_summary(colors_chosen)
            break

    print("\n🌠 May the stars shine upon you, farewell! 🌙✨")

# 🔥 START THE GAME
fortune_teller()
```
