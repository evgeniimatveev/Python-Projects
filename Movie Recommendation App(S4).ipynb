```python
# Movie Recommendation App
# This app provides movie suggestions based on the user's age, along with trending movie recommendations.

# Step 1: Define a function to recommend a movie based on the user's age
def recommend_movie(age):
    """
    Recommends a movie based on the user's age.
    
    Parameters:
        age (int): The age of the user.
    Returns:
        str: A detailed movie recommendation.
    """
    # Recommendations for different age ranges
    if age < 3:
        return (
            "🎥 **Recommended Movie:** Baby Shark's Big Show 🦈\n"
            "📝 **Summary:** An engaging and fun show perfect for toddlers with musical adventures!"
        )
    elif 3 <= age < 6:
        return (
            "🎥 **Recommended Movie:** Peppa Pig 🐷\n"
            "📝 **Summary:** Lighthearted stories about Peppa and her family. A favorite among preschoolers."
        )
    elif 6 <= age < 9:
        return (
            "🎥 **Recommended Movie:** Toy Story 🎠\n"
            "📝 **Summary:** A heartwarming story about toys coming to life, with themes of friendship and adventure."
        )
    elif 9 <= age < 12:
        return (
            "🎥 **Recommended Movie:** Inside Out 🎭\n"
            "📝 **Summary:** A touching and educational movie exploring emotions with vivid characters."
        )
    elif 12 <= age < 15:
        return (
            "🎥 **Recommended Movie:** Spider-Man: Into the Spider-Verse 🕸️\n"
            "📝 **Summary:** A visually stunning superhero movie with a great story about courage and teamwork."
        )
    elif 15 <= age < 18:
        return (
            "🎥 **Recommended Movie:** Black Panther 🐾\n"
            "📝 **Summary:** A cultural phenomenon with powerful storytelling and action-packed sequences."
        )
    elif 18 <= age < 21:
        return (
            "🎥 **Recommended Movie:** Joker 🃏\n"
            "📝 **Summary:** A deep psychological drama exploring the transformation of a man into the iconic Joker."
        )
    elif 21 <= age < 24:
        return (
            "🎥 **Recommended Movie:** Parasite 🎞️\n"
            "📝 **Summary:** A genre-bending masterpiece that combines comedy, suspense, and social commentary."
        )
    elif 24 <= age < 27:
        return (
            "🎥 **Recommended Movie:** Interstellar 🌌\n"
            "📝 **Summary:** A visually breathtaking sci-fi film exploring love, time, and humanity's survival."
        )
    elif 27 <= age < 30:
        return (
            "🎥 **Recommended Movie:** The Matrix 🟩\n"
            "📝 **Summary:** A groundbreaking sci-fi thriller about the nature of reality."
        )
    elif 30 <= age < 40:
        return (
            "🎥 **Recommended Movie:** The Godfather 👑\n"
            "📝 **Summary:** A classic mafia drama exploring power, family, and betrayal."
        )
    elif 40 <= age < 50:
        return (
            "🎥 **Recommended Movie:** Schindler's List 📜\n"
            "📝 **Summary:** A deeply moving historical drama about heroism during the Holocaust."
        )
    elif 50 <= age < 60:
        return (
            "🎥 **Recommended Movie:** Forrest Gump 🏃‍♂️\n"
            "📝 **Summary:** A heartfelt story about an extraordinary man's journey through life's challenges."
        )
    elif 60 <= age < 70:
        return (
            "🎥 **Recommended Movie:** The Sound of Music 🎼\n"
            "📝 **Summary:** A timeless musical about love, family, and resilience during World War II."
        )
    elif 70 <= age < 81:
        return (
            "🎥 **Recommended Movie:** Casablanca 🕊️\n"
            "📝 **Summary:** A romantic classic set during World War II, featuring unforgettable performances."
        )
    elif age >= 81:
        return (
            "🎥 **Recommended Movie:** It's a Wonderful Life 🎄\n"
            "📝 **Summary:** A feel-good holiday classic about appreciating life's true value."
        )
    else:
        return "❌ Sorry, no recommendations are available for this age range."

# Step 2: Add a trending movies feature
def trending_movies():
    """
    Displays a list of trending movies for all ages.
    """
    trending_list = [
        "🎥 Avatar: The Way of Water 🌊",
        "🎥 Top Gun: Maverick ✈️",
        "🎥 Everything Everywhere All At Once 🌌",
        "🎥 Dune 🏜️",
        "🎥 Encanto ✨"
    ]
    print("🔥 **Trending Movies**:")
    for movie in trending_list:
        print(f"- {movie}")

# Step 3: Create the main function
def movie_app():
    """
    Main function for interacting with the user.
    """
    print("🎬 Welcome to the Movie Recommendation App!")
    print("🔍 Discover the perfect movie for your age and explore trending titles.")
    print("Type -1 at any time to quit. 🛑")
    
    # Display trending movies
    trending_movies()
    print("\n")

    while True:
        try:
            # Ask for user's age
            age = int(input("Enter your age (or type -1 to quit): "))
            if age == -1:
                print("Thank you for using the app. Goodbye! 😊")
                break
            # Provide recommendations
            print(recommend_movie(age))
        except ValueError:
            print("❌ Please enter a valid age.")

# Step 4: Run the app
movie_app()
```
