Welcome to the **intro Jupyter notebook** for the Python Projects repository.

In this notebook you will:
- See a quick overview of the available project branches
- Load them into a small Pandas DataFrame
- Sort projects by difficulty level

Use this as a starting point before exploring individual project branches.


```python
# Quick overview of projects using Pandas

import pandas as pd

projects = [
    ("avocado-prices", "Avocado Prices Prediction", "ML / Time Series", "intermediate"),
    ("amazon-reviews-nlp", "Amazon Reviews Sentiment", "NLP", "intermediate"),
    ("realtime-ai-filters", "Real-Time AI Filters", "Computer Vision", "advanced"),
    ("fortune-teller", "Fortune Teller Game", "Game / Logic", "beginner"),
    ("guessing-game", "Number Guessing Game", "Game", "beginner"),
    ("image-processing-app", "Image Processing App", "OpenCV", "intermediate"),
    ("zodiac-finder", "Zodiac Sign Finder", "Utility", "beginner"),
    ("tic-tac-toe-ai", "Tic Tac Toe with simple AI", "Game / AI", "intermediate"),
    ("opencv-realtime", "OpenCV Real-time Processing", "Computer Vision", "advanced"),
    ("movie-recommender", "Movie Recommendation App", "Recommender System", "intermediate"),
]

df = pd.DataFrame(projects, columns=["branch", "title", "topic", "level"])
df
```

```python
# Sort projects by difficulty level

level_order = ["beginner", "intermediate", "advanced"]
df["level"] = pd.Categorical(df["level"], categories=level_order, ordered=True)

df_sorted = df.sort_values("level")
df_sorted
```
