```python
import tkinter as tk
from tkinter import messagebox
import datetime
import itertools
import random  
from reportlab.pdfgen import canvas  # âœ… PDF support

# ğŸ¨ Dynamic Background Animation (Gradient Effect)
def animate_bg():
    """ğŸŒˆ Creates a smooth gradient effect that shifts between colors."""
    colors = itertools.cycle(["#222831", "#2a3137", "#32383e", "#3a4046", "#444b52", "#222831"])  
    def change_color():
        root.config(bg=next(colors))
        root.after(1000, change_color)  
    change_color()

# ğŸ”¥ Zodiac Sign Calculation
def get_zodiac_sign(day, month):
    """Returns the zodiac sign based on birth date."""
    zodiac_dates = [
        ("Capricorn", 12, 22, 1, 19), ("Aquarius", 1, 20, 2, 18), ("Pisces", 2, 19, 3, 20),
        ("Aries", 3, 21, 4, 19), ("Taurus", 4, 20, 5, 20), ("Gemini", 5, 21, 6, 20),
        ("Cancer", 6, 21, 7, 22), ("Leo", 7, 23, 8, 22), ("Virgo", 8, 23, 9, 22),
        ("Libra", 9, 23, 10, 22), ("Scorpio", 10, 23, 11, 21), ("Sagittarius", 11, 22, 12, 21),
        ("Capricorn", 12, 22, 1, 19)
    ]
    for sign, start_month, start_day, end_month, end_day in zodiac_dates:
        if (month == start_month and day >= start_day) or (month == end_month and day <= end_day):
            return sign
    return "Unknown"

# ğŸ”¢ Numerology Calculation
def calculate_numerology_number(year):
    """Returns a single-digit numerology number based on birth year."""
    while year >= 10:
        year = sum(int(digit) for digit in str(year))
    return year

# ğŸ”® Extra Magic Features
def get_lucky_element(zodiac_sign):
    elements = {"Fire": ["Aries", "Leo", "Sagittarius"], "Earth": ["Taurus", "Virgo", "Capricorn"],
                "Air": ["Gemini", "Libra", "Aquarius"], "Water": ["Cancer", "Scorpio", "Pisces"]}
    for element, signs in elements.items():
        if zodiac_sign in signs:
            return element
    return "Unknown"

def get_lucky_numbers(name):
    """Generates 3 lucky numbers based on name."""
    random.seed(sum(ord(char) for char in name))
    return random.sample(range(1, 100), 3)

def get_lucky_color(zodiac_sign):
    colors = {"Aries": "Red â¤ï¸", "Taurus": "Green ğŸ’š", "Gemini": "Yellow ğŸ’›",
              "Cancer": "White ğŸ¤", "Leo": "Gold âœ¨", "Virgo": "Brown ğŸ¤",
              "Libra": "Pink ğŸ€", "Scorpio": "Black ğŸ–¤", "Sagittarius": "Purple ğŸ’œ",
              "Capricorn": "Grey âš«", "Aquarius": "Blue ğŸ’™", "Pisces": "Sea Green ğŸŒŠ"}
    return colors.get(zodiac_sign, "Unknown")

def get_mystical_message():
    messages = ["âœ¨ The universe is aligning in your favor...", "ğŸŒ™ Trust your intuition, it will guide you to success.",
                "ğŸ”® A significant change is coming, embrace it with confidence.",
                "ğŸ”¥ Your inner fire is stronger than you think, use it wisely.",
                "ğŸŒ¿ A moment of calm will bring unexpected clarity."]
    return random.choice(messages)

# ğŸ“œ Report Formatting
def format_report(name, day, month, year, zodiac_sign, num_number):
    lucky_element = get_lucky_element(zodiac_sign)
    lucky_numbers = get_lucky_numbers(name)
    lucky_color = get_lucky_color(zodiac_sign)
    mystical_message = get_mystical_message()
    
    return f"""
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
ğŸ”® ğ—¬ğ—¢ğ—¨ğ—¥ ğ—”ğ—¦ğ—§ğ—¥ğ—¢ ğ—¥ğ—˜ğ—£ğ—¢ğ—¥ğ—§ ğŸ”®
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
ğŸ‘¤ Name: {name}
ğŸ‚ Birthdate: {day}/{month}/{year} (Age: {datetime.date.today().year - year})
âœ¨ Zodiac Sign: {zodiac_sign}
ğŸ”¢ Numerology Number: {num_number}
ğŸ”¥ Lucky Element: {lucky_element}
ğŸ¨ Lucky Color: {lucky_color}
ğŸ² Lucky Numbers: {', '.join(map(str, lucky_numbers))}
ğŸŒ  Mystical Message: "{mystical_message}"
May the stars guide your path! âœ¨
"""

# ğŸ’¾ Save Report as PDF
def save_report():
    """Save the astrology report as a PDF."""
    text = result_label.cget("text")
    
    pdf = canvas.Canvas("Astro_Report.pdf")
    pdf.setFont("Helvetica", 12)

    y_position = 800  
    for line in text.split("\n"):
        pdf.drawString(50, y_position, line)
        y_position -= 20  

    pdf.save()
    messagebox.showinfo("Success", "Report saved as 'Astro_Report.pdf'")

# ğŸ”® Generate Report
def generate_report():
    name = name_entry.get()
    try:
        day = int(day_entry.get())
        month = int(month_entry.get())
        year = int(year_entry.get())
    except ValueError:
        messagebox.showerror("Input Error", "Please enter valid numbers for date!")
        return

    if not name:
        messagebox.showerror("Input Error", "Please enter your name!")
        return

    zodiac_sign = get_zodiac_sign(day, month)
    num_number = calculate_numerology_number(year)

    result_label.config(text=format_report(name, day, month, year, zodiac_sign, num_number), justify="center")

# ğŸš€ GUI Setup
root = tk.Tk()
root.title("Zodiac & Numerology Analyzer")
root.geometry("550x700")
animate_bg()  

# ğŸŒŸ Glowing Title
title_label = tk.Label(root, text="ğŸ”® Enter Your Birth Info", font=("Arial", 18, "bold"), fg="#FFD700", bg="#222831")
title_label.pack(pady=15)

# ğŸ“… Create input fields
def create_label(text):
    return tk.Label(root, text=text, font=("Arial", 12), fg="white", bg="#222831", anchor="center")

create_label("Name:").pack()
name_entry = tk.Entry(root, font=("Arial", 12), width=25, justify="center")
name_entry.pack()

create_label("Day (1-31):").pack()
day_entry = tk.Entry(root, font=("Arial", 12), width=25, justify="center")
day_entry.pack()

create_label("Month (1-12):").pack()
month_entry = tk.Entry(root, font=("Arial", 12), width=25, justify="center")
month_entry.pack()

create_label("Year:").pack()
year_entry = tk.Entry(root, font=("Arial", 12), width=25, justify="center")
year_entry.pack()

# ğŸ­ Buttons
generate_btn = tk.Button(root, text="âœ¨ Generate Report", command=generate_report, bg="#00cc99", fg="white", font=("Arial", 14, "bold"))
generate_btn.pack(pady=10)

save_btn = tk.Button(root, text="ğŸ’¾ Save as PDF", command=save_report, bg="#ff9966", fg="white", font=("Arial", 14, "bold"))
save_btn.pack(pady=5)

# ğŸ“œ Display frame
result_label = tk.Label(root, text="", wraplength=450, font=("Arial", 12), justify="center", fg="white", bg="#393E46")
result_label.pack()

root.mainloop()
```
